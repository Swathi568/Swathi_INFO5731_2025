The formula that describes the volume V of a hemisphere with radius r is the following:

$ V=\frac{2}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of even numbers from 1-20. Output the result in the following format:

	r	V
	2	…
	4	…
	6	…
	.
	.
	.
	20	…


```python
import math

print("r\tV")  #assigning columns

for r in range(1, 21):  #range of r from 1 to 20
    if r % 2 == 0:  #checking r is even or not
        V = (2/3) * math.pi * (r**3)  # math formula
        print(f"{r}\t{V:.2f}")  # v with two decimal points


```

(b) Transform each sentence into a list, where each element corresponds to a word in the sentence. For example, if the sentences are "Hello, how are you doing today?" and "This is an example sentence for the program!"

The program should output:
['Hello', ',', 'how', 'are', 'you', 'doing', 'today', '?', 'This', 'is', 'an', 'example', 'sentence', 'for', 'the', 'program', '!']

(c) Compare the two sentences. Print out a list of words that occur in both sentences.

(d) Remove any punctuation from the sentence and print the sanitized sentences. It should remove at least the following characters, but it can remove more: period(“.”), comma (“,”), semicolon (“;”), and colon (“:”).

(e) Count the number of vowels in each of the sentences. Print out each vowel and the occurrences  in the each of the two sentences separately, such as:

Sentence1: a:2, e:2, i:0, etc.
Sentence2: a:1, e:2, i:2, etc.

Design a well-structured Python program to execute these tasks efficiently.

```python

import string  # for punctuation handling

# (a)
sentence_1 = input("Enter the 1st sentence: ")
sentence_2 = input("Enter the 2nd sentence: ")

# (b)
#Converting sentences into lists
sentence_1_list = sentence_1.split()
sentence_2_list = sentence_2.split()
combined_list = sentence_1_list + sentence_2_list

print("\nList_of_words_from_both_sentences:")
print(combined_list)

# (c) finding common words
common_words = list(set(sentence_1_list) & set(sentence_2_list))
print("\nCommon_words_in_both_sentences:")
print(common_words if common_words else "No common words found.")

# (d) Removing punctuation
def remove_punctuation(text):
    return text.translate(str.maketrans("", "", string.punctuation))

sentence_1_clean = remove_punctuation(sentence_1)
sentence_2_clean = remove_punctuation(sentence_2)

print("\nSentences(Without Punctuation):")
print("Sentence_1:", sentence_1_clean)
print("Sentence_2:", sentence_2_clean)

# (e) Counting vowels in each sentence
def count_vowels(text):
    vowels = "aeiou"
    text = text.lower()
    return {v: text.count(v) for v in vowels}

vowel_count_1 = count_vowels(sentence_1_clean)
vowel_count_2 = count_vowels(sentence_2_clean)

print("\nVowel_occurrences_in_Sentence_1:")
print(", ".join([f"{v}:{c}" for v, c in vowel_count_1.items()]))

print("\nVowel_occurrences_in_Sentence_2:")
print(", ".join([f"{v}:{c}" for v, c in vowel_count_2.items()]))

```

(d) Implement error handling in the calculator program. Handle cases such as division by zero and invalid input expressions. Provide meaningful error messages.

```python
import math

def evaluate_expression(expression):
    try:
        result = eval(expression)
        return result
    except ZeroDivisionError:
        return "Error: Division by zero is not valid."
    except Exception as e:
        return f"Error: {str(e)}"

def calculator():
    print("Welcome to the Calculator!")
    print("operations_supported: +, -, *, /, ** (exponentiation), % (modulus), sqrt (square root)")
    print("Enter 'exit' to quit the calculator.")

    while True:
        expression = input("\nEnter a mathematical expression: ")

        if expression.lower() == 'exit':
            print("Exiting the calculator!")
            break

        #square root
        expression = expression.replace("sqrt", "math.sqrt")

        result = evaluate_expression(expression)

        print(f"Result: {result}")

# Run the calculator
calculator()
```

                              Year	Value
	Vacant Housing Units	  -----   -----
	Occupied Housing Units	-----   -----

(d) (5 pts) Calculate the harmonic mean of the total housing units and print it out. You can find the information about harmonic mean here: https://ncalculators.com/statistics/harmonic-mean-calculator.htm

```python

import pandas as pd
from scipy.stats import hmean

# Loading file
df = pd.read_csv('Assignment1_denton_housing.csv')

# Filtering the data
occupied = df[df['title_field'] == 'Occupied Housing Units']
vacant = df[df['title_field'] == 'Vacant Housing Units']

# (a)
occupied = occupied.sort_values(by='year')
occupied['difference'] = occupied['value'].diff()
print(occupied[['year', 'difference']].dropna().sort_values(by='difference'))

# (b)
merged = pd.merge(occupied, vacant, on='year', suffixes=('_occupied', '_vacant'))
merged['percent_occupied'] = (merged['value_occupied'] / (merged['value_occupied'] + merged['value_vacant'])) * 100
merged['percent_vacant'] = (merged['value_vacant'] / (merged['value_occupied'] + merged['value_vacant'])) * 100
print("Year  Vacant   Housing Units    Occupied Housing Units")
for index, row in merged.iterrows():
    print(f"{int(row['year'])}   {row['percent_vacant']:.0f}%   {row['percent_occupied']:.0f}%")

#  (c)
max_vacant = vacant.loc[vacant['value'].idxmax()]
max_occupied = occupied.loc[occupied['value'].idxmax()]
print("                          Year    Value")
print(f"Vacant Housing Units      {int(max_vacant['year'])}   {max_vacant['value']}")
print(f"Occupied Housing Units    {int(max_occupied['year'])}   {max_occupied['value']}")

#(d)
total_housing = df[df['title_field'] == 'Total Housing Units']
harmonic_mean = hmean(total_housing['value'])
print(f"Harmonic Mean of Total Housing Units: {harmonic_mean:.2f}")
```
